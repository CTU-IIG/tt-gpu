**General experiment description**
For this experiments up to 16 kernels were launched in different streams. This leads to a parallel execution and therefore contention between the kernels on memory level. For each run of the experiment the kernels executed the sequential walk on a different sized array and with changing number of threads. If multiple threads are used within one kernel the sequential walk is performed in a strided manner:


| Thread0 | Thread1 | Thread2 | Thread3 | ...     | Thread0      | Thread1    | Thread2    | Thread3    |...|
| ---     | ---     | ---     | ---     | ---     | ---          | ---        | ---        | ---        | ---|
| 0       | 1       | 2       | 4       | ...     | 0+blockDim.x |1+blockDim.x|2+blockDim.x|3+blockDim.x|...|


In parallel to this contention, the CPU were utilized to either perform random or sequential memory copy between two arrays of size 8MB each.

**Results**
On the first sight, it can be seen that the sequential interference adds much more contention on the memory then the random interference. This might indicate that the memory controller priorizes sequential accesses to memory over random accesses.

Additionally the L2 cache contention between kernels is shown in both figures. If the random walks with 1 thread are excluded, it can be seen, that for the 512MB dataset size the contention starts already with two kernels in parallel, whereas the contention for 256MB and 128MB datasets starts with three respectively with five kernels in parallel. After this number of parallel executing kernels evictions of data which was prefetched by another kernel is getting more and more probable. The fact that the kernels running only 32 respectively 1 thread suffer of less contention is supported by the random replacement policy of L2 GPU cache. This means if more threads are accessing L2 the probability of evictions increases.
